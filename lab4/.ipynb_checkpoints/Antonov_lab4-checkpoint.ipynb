### Задача 4.1.2

Найти с точностью  $\mathcal{E} = 10^{-6}$ все корни системы нелинейных уравнений:
$$f_1(x_1,x_2) =0$$
$$f_2(x_1,x_2) =0$$

используя метод Ньютона для системы нелинейных уравнений.

$f_1=cos(x_1-1)+x_2-0.5$

$f_2=sin(x_1)+2 x_2-2$

```python
def funcf1(x1,x2):
    return np.cos(x1-1)+x2-0.5

def funcf2(x1,x2):
    return np.sin(x1)+2*x2-2

def functions(xy):
    x1, x2 = xy
    return np.array([np.cos(x1-1)+x2-0.5, np.sin(x1)+2*x2-2])
```

```python
def func_g1_x2(x1):
    return -np.cos(x1-1)+0.5

def func_g2_x2(x1):
    return -0.5*np.sin(x1)+1
```

#### 1)Локализация корней

```python
x = np.linspace(-10, 10, 500)

dotX = [3.0,-1.90,-3.2,4.3]
dotY = [0.94,1.47,0.94,1.47]

fig, ax = plt.subplots()
ax.plot(x, func_g1_x2(x),label = "g1_x2")
ax.plot(x, func_g2_x2(x),label = "g2_x2")

ax.legend()

plt.scatter(dotX,dotY,color='red')
ax = plt.gca()
ax.axhline(y=0, color='black')
ax.axvline(x=0, color='black');
```

#### 2)  Метод Ньютона

```python
def newton_system(f, Df, x0, tol=1e-6, maxiter=100):
    """
    Решение систем нелинейных уравнений методом Ньютона
    
    Параметры:
    f - функция, возвращающая значения системы уравнений
    Df - функция, возвращающая якобиан системы уравнений
    x0 - начальное приближение
    tol - требуемая точность
    maxiter - максимальное число итераций
    
    Возвращает:
    x - решение системы уравнений
    """
    
    # итерационный процесс
    for i in range(maxiter):
        # вычисляем значения функции и якобиана в текущей точке
        f_val = f(x0)
        Df_val = Df(x0)

        # решаем систему линейных уравнений для определения приращения
        delta = np.linalg.solve(Df_val, -f_val)

        # обновляем значение x0
        x0 += delta

        # проверяем на достижение заданной точности
        if np.linalg.norm(delta) < tol:
            return x0, i
    print('Превышено число итераций')
    return None, i

def jacobian(xy):
    x_1, x_2 = xy
    return np.array([[-np.sin(x_1 - 1),1],
            [np.cos(x_1), 2]])
```

#### 3) вычисление корней

Вычислим два корня, так как функции периодичны остальные корни будут смещены на $2\pi$

```python
root, iterations = newton_system(functions,jacobian,[3.03, 0.9])
print(f'x1 = {root[0]}, x2 = {root[1]}, итерации = {iterations}')

root, iterations = newton_system(functions,jacobian,[4, 5])
print(f'x1 = {root[0]}, x2 = {root[1]}, итерации = {iterations}')
```

#### 4) поиск корней встроенной функцией

```python
root = scp.root(functions, [3.03, 0.9], tol=1e-6)
print(f'x1 = {root.x[0]}, x2 = {root.x[1]}')

root = scp.root(functions, [4, 5], tol=1e-6)
print(f'x1 = {root.x[0]}, x2 = {root.x[1]}')
```

### 4.3.1

Найти с точностью  $\mathcal{E} = 10^{-6}$ все корни системы нелинейных уравнений:
$$f_1(x_1,x_2) =0$$
$$f_2(x_1,x_2) =0$$

используя метод простой итерации для системы нелинейных уравнений. Проверить выполнение достаточного
условия сходимости метода (использовать норму
||$\cdot $||$_\infty$

$f_1=0.7x_1-(sin(x_2))/3-2$

$f_2=1.1x_1+2x_2-sin(x_1/5)+1$

```python
def functions2(xy):
    x1, x2 = xy
    return np.array([0.7 * x1 - (np.sin(x2))/3 - 2, 1.1 * x1 + 2 * x2 - np.sin(x1/5) + 1])
```

#### 1) Локализуем корни

```python
def func_g1(x2):
    return np.sin(x2)/(3*0.7)+2/0.7

def func_g2(x1):
    return -1.1*x1/2 + np.sin(x1/5)/2 - 0.5

def mainFunc(xy):
    x1, x2 = xy
    
    x1 = func_g1(x2)
    x2 = func_g2(x1)
    return np.array([x1,x2])
```

```python
x1 = np.linspace(-2, 3, 500)
x2 = np.linspace(-2, 3, 500)

dotX = [2.380]
dotY = [-1.58]

fig, ax = plt.subplots()
ax.plot(func_g1(x2),x2,label = "g1_x2")
ax.plot(x1, func_g2(x1),label = "g2_x2")

ax.legend()

plt.scatter(dotX,dotY,color='red')
ax = plt.gca()
ax.axhline(y=0, color='black')
ax.axvline(x=0, color='black');
```

#### 2) реализуем метод простой итерации

```python
def J(xy):
    x_1, x_2 = xy
    return np.array([[0,np.cos(x_2)/(3*0.7)],
            [-1.1/2 + np.cos(x_1/5)/10,0]])
```

```python
import numpy as np

def simple_iteration(F, x0, tol=1e-6, max_iter=100):
    """
    Решает систему нелинейных уравнений методом простой итерации.

    Аргументы:
        - F: функция, принимающая на вход вектор x и возвращающая вектор значений F(x).
        - x0: начальное приближение для решения системы нелинейных уравнений.
        - tol: требуемая точность решения.
        - max_iter: максимальное число итераций.

    Возвращает:
        - x: приближенное решение системы нелинейных уравнений.
    """
    x = np.array(x0)
    q = np.linalg.norm(J(x), np.inf)
    for i in range(max_iter):
        Fx = F(x)
        norm_Fx = np.linalg.norm(Fx-x, ord=np.inf)
        if norm_Fx < (1 - q) * tol / q:
            if 0 <= q < 1:
                print('Достаточное условие выполнено')
            else:
                print('Достаточное условие не выполнено')
            return x,i
        x = Fx
    return x,i
```

```python
x_0 = [2, -1]
root, iterations = simple_iteration(mainFunc,x_0)
print(f'x1 = {root[0]}, x2 = {root[1]}, итерации = {iterations}')
```
