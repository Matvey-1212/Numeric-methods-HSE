–ü–û–†–Ø–î–û–ö –†–ï–®–ï–ù–ò–Ø –ó–ê–î–ê–ß–ò:
1. –ó–∞–¥–∞—Ç—å –º–∞—Ç—Ä–∏—Ü—É —Å–∏—Å—Ç–µ–º—ã A –∏ –≤–µ–∫—Ç–æ—Ä –ø—Ä–∞–≤–æ–π —á–∞—Å—Ç–∏ b. –ò—Å–ø–æ–ª—å–∑—É—è –≤—Å—Ç—Ä–æ–µ–Ω–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é lsolve –ø–∞–∫–µ—Ç–∞ MATHCAD (–∏–ª–∏ –¥—Ä—É–≥–∏–µ –≤—Å—Ç—Ä–æ–µ–Ω–Ω—ã–µ —Ñ—É–Ω–∫—Ü–∏–∏ –≤ wolfram mathematica –∏–ª–∏ –≤ python), –Ω–∞–π—Ç–∏ —Ä–µ—à–µ–Ω–∏–µ —Å–∏—Å—Ç–µ–º—ã $Ax=b$ —Å –ø–æ–º–æ—â—å—é –º–µ—Ç–æ–¥–∞ –ì–∞—É—Å—Å–∞.
2. –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞—Ç—å —Å–∏—Å—Ç–µ–º—É $Ax=b$ –∫ –≤–∏–¥—É $x=Bx+c$, —É–¥–æ–±–Ω–æ–º—É –¥–ª—è –∏—Ç–µ—Ä–∞—Ü–∏–π. –ü—Ä–æ–≤–µ—Ä–∏—Ç—å –≤—ã–ø–æ–ª–Ω–µ–Ω–∏–µ –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ–≥–æ
—É—Å–ª–æ–≤–∏—è —Å—Ö–æ–¥–∏–º–æ—Å—Ç–∏ –∏—Ç–µ—Ä–∞—Ü–∏–æ–Ω–Ω—ã—Ö –º–µ—Ç–æ–¥–æ–≤ $||B||_{\inf}<1$.
3. –ù–∞–ø–∏—Å–∞—Ç—å –ø—Ä–æ–≥—Ä–∞–º–º—É, —Ä–µ—à–∞—é—â—É—é —Å–∏—Å—Ç–µ–º—É –ª–∏–Ω–µ–π–Ω—ã—Ö —É—Ä–∞–≤–Ω–µ–Ω–∏–π —Å –ø–æ–º–æ—â—å—é –º–µ—Ç–æ–¥–∞ –ó–µ–π–¥–µ–ª—è, –≤—ã–ø–æ–ª–Ω–∏—Ç—å 10 –∏—Ç–µ—Ä–∞—Ü–∏–π –ø–æ –º–µ—Ç–æ–¥—É –ó–µ–π–¥–µ–ª—è; –≤–∑—è—Ç—å –ª—é–±–æ–µ –Ω–∞—á–∞–ª—å–Ω–æ–µ –ø—Ä–∏–±–ª–∏–∂–µ–Ω–∏–µ. –ü—Ä–µ–¥–≤–∞—Ä–∏—Ç–µ–ª—å–Ω–æ –ø—Ä–æ–≤–µ—Ä–∏—Ç—å —É—Å–ª–æ–≤–∏–µ —Å—Ö–æ–¥–∏–º–æ—Å—Ç–∏ –¥–ª—è –º–µ—Ç–æ–¥–∞ –ó–µ–π–¥–µ–ª—è. –ü—Ä–∏–Ω–∏–º–∞—è —Ä–µ—à–µ–Ω–∏–µ, –ø–æ–ª—É—á–µ–Ω–Ω–æ–µ –≤ –ø. 1 –∑–∞ —Ç–æ—á–Ω–æ–µ, –Ω–∞–π—Ç–∏ –≤–µ–ª–∏—á–∏–Ω—É –∞–±—Å–æ–ª—é—Ç–Ω–æ–π
–ø–æ–≥—Ä–µ—à–Ω–æ—Å—Ç–∏ –∏—Ç–µ—Ä–∞—Ü–∏–æ–Ω–Ω–æ–≥–æ —Ä–µ—à–µ–Ω–∏—è (–∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –Ω–æ—Ä–º—É $|| \times ||_{\inf}$).
4. –í–∑—è—Ç—å –¥—Ä—É–≥–æ–µ –Ω–∞—á–∞–ª—å–Ω–æ–µ –ø—Ä–∏–±–ª–∏–∂–µ–Ω–∏–µ. –û–±—ä—è—Å–Ω–∏—Ç—å –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã.

```python
A = np.array([[29.7, 2., 0., 19.8, 2.],
              [9.9, -21., 0., -9.9, 1.],
              [-9.9, 11., 29., 6.6, 1.],
              [9.9, 7.5, 2., -19.8, 0.],
              [-49.5, -1., 23., 9.9, 84.],
             ])
b = np.array([26.2, -41.1, 97.4, 99.8, 27.1])
```

#### –ù–∞–π–¥–µ–º —Ä–µ—à–µ–Ω–∏–µ —Å–∏—Å—Ç–µ–º—ã –≤—Å—Ç—Ä–æ–µ–Ω–Ω—ã–º –º–µ—Ç–æ–¥–æ–º

```python
x_solve1 = np.linalg.solve(A, b)
for i in range(len(x_solve1)):
    print("x",i+1,' = ', x_solve1[i],sep='')
```

#### –ü—Ä–∏–≤–¥–µ–º $ùê¥ùë•=ùëè$ –∫ –≤–∏–¥—É  $ùë•=ùêµùë•+ùëê$

```python
B = np.zeros_like(A)
c = np.zeros_like(b)

for i in range(len(B)):
    for j in range(len(B)):
        if i != j:
            B[i][j] = -A[i][j]/A[i][i]
    c[i] = b[i]/A[i][i]
print(B)
print()
print(c)
```

–ü—Ä–æ–≤–µ—Ä–∏–º —Å—Ö–æ–¥–∏–º–æ—Å—Ç—å

```python
np.linalg.norm(B, np.inf)
```

–ú–µ–Ω—å—à–µ 1, –∑–Ω–∞—á–∏—Ç –º–µ—Ç–æ–¥ –ó–µ–π–¥–µ–ª—è —Å—Ö–æ–¥–∏—Ç—Å—è

### –ú–µ—Ç–æ–¥–æ –ó–µ–π–¥–µ–ª—è

```python
def zeidel(B, c, it, x0):
    n = len(B)
    B1 = np.tri(n)*B
    B2 = (np.tri(n).T)*B
    for i in range(it):
        xk = B@x0 +c
        x0 = B1@xk + B2@xk + c
    return xk
```

```python
x0=[0.,0.,0.,0.,0.]
x_z = zeidel(B, c, 10, x0=x0)
print('–†–µ—à–µ–Ω–∏–µ –ø—Ä–∏ –ø—Ä–∏–±–ª–∏–∂–µ–Ω–∏–∏:', x0)
for i in range(len(x_z)):
    print("x",i+1,' = ', x_z[i],sep='')
abs_err = np.linalg.norm(x_z - x_solve1, np.inf)
print('–ê–±—Å–æ–ª—é—Ç–Ω–∞—è –ø–æ–≥—Ä–µ—à–Ω–æ—Å—Ç—å: ', abs_err)
```

### –ó–∞–¥–∞—á–∞ 5.2

–î–ª—è —Å–∏—Å—Ç–µ–º—ã —É—Ä–∞–≤–Ω–µ–Ω–∏–π Ax=b –∏–∑ –∑–∞–¥–∞—á–∏ 5.1 –Ω–∞–π—Ç–∏ —Ä–µ—à–µ–Ω–∏–µ –ø–æ –º–µ—Ç–æ–¥—É –ó–µ–π–¥–µ–ª—è —Å —Ç–æ—á–Ω–æ—Å—Ç—å—é
$\epsilon=10^{-6}$ , –≤–∑—è–≤ –ª—é–±–æ–µ –Ω–∞—á–∞–ª—å–Ω–æ–µ –ø—Ä–∏–±–ª–∏–∂–µ–Ω–∏–µ. –î–ª—è —ç—Ç–æ–≥–æ –º–æ–¥–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞—Ç—å –ø—Ä–æ–≥—Ä–∞–º–º—É –∏–∑ –ø—É–Ω–∫—Ç–∞ 3 —Ç–∞–∫, —á—Ç–æ–±—ã —Ä–µ—à–µ–Ω–∏–µ –≤—ã—á–∏—Å–ª—è–ª–æ—Å—å —Å –∑–∞–¥–∞–Ω–Ω–æ–π —Ç–æ—á–Ω–æ—Å—Ç—å—é $\epsilon$. –ü—Ä–µ–¥—É—Å–º–æ—Ç—Ä–µ—Ç—å –ø–æ–¥—Å—á–µ—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∏—Ç–µ—Ä–∞—Ü–∏–π, –ø–æ—Ç—Ä–µ–±–æ–≤–∞–≤—à–∏—Ö—Å—è –¥–ª—è –¥–æ—Å—Ç–∏–∂–µ–Ω–∏—è —Ç–æ—á–Ω–æ—Å—Ç–∏ $\epsilon$.

```python
def zeidel_tol(B, c, tol=1e-6, it=100, x0=[0.,0.,0.,0.,0.]):
    n = len(B)
    B1 = np.tri(n)*B
    B2 = (np.tri(n).T)*B
    iterations = 0
    x0 = np.ones(n)
    xk = np.copy(x0)
    while True: 
        iterations+=1
        xk = B@x0 +c
        x0 = B1@xk + B2@xk + c
        if iterations > it:
            print(f"–ü—Ä–µ–≤—ã—à–µ–Ω–æ –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–µ –Ω–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏—Ç–µ—Ä–∞—Ü–∏–π ({it})")
            return xk,it
        if (np.linalg.norm(x0 - xk, np.inf) < tol):
            return xk,iterations
    return xk,iterations
```

```python
tol=1e-6
x_z, iters = zeidel_tol(B, c, tol=tol)
print('–†–µ—à–µ–Ω–∏–µ —Å –ø–æ–≥—Ä–µ—à–Ω–æ—Å—Ç—å—é:',tol)
for i in range(len(x_z)):
    print("x",i+1,' = ', x_z[i],sep='')
print('–ö–æ–ª–≤–æ –∏—Ç–µ—Ä–∞—Ü–∏–π',iters)
print()
abs_err = np.linalg.norm(x_z - x_solve1, np.inf)
print('–ê–±—Å–æ–ª—é—Ç–Ω–∞—è –ø–æ–≥—Ä–µ—à–Ω–æ—Å—Ç—å: ', abs_err)
```

### –ó–∞–¥–∞—á–∞ 5.7.1

–î–∞–Ω–∞ —Å–∏—Å—Ç–µ–º–∞ —É—Ä–∞–≤–Ω–µ–Ω–∏–π $Ax=b$, –≥–¥–µ A ‚Äì —Å–∏–º–º–µ—Ç—Ä–∏—á–Ω–∞—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞
—Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏ $n —Ö n$. –ù–∞–π—Ç–∏ —Ä–µ—à–µ–Ω–∏–µ —Å–∏—Å—Ç–µ–º—ã —Å —Ç–æ—á–Ω–æ—Å—Ç—å—é $\epsilon=10^{-6}$ —Å –ø–æ–º–æ—â—å—é –º–µ—Ç–æ–¥–∞ –ø—Ä–æ—Å—Ç–æ–π –∏—Ç–µ—Ä–∞—Ü–∏–∏ —Å –æ–ø—Ç–∏–º–∞–ª—å–Ω—ã–º –≤—ã–±–æ—Ä–æ–º –∏—Ç–µ—Ä–∞—Ü–∏–æ–Ω–Ω–æ–≥–æ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞, –∞ —Ç–∞–∫–∂–µ —Å –ø–æ–º–æ—â—å—é –º–µ—Ç–æ–¥–∞, —É–∫–∞–∑–∞–Ω–Ω–æ–≥–æ –≤ –∏–Ω–¥–∏–≤–∏–¥—É–∞–ª—å–Ω–æ–º –≤–∞—Ä–∏–∞–Ω—Ç–µ. (–î–ª—è –º–µ—Ç–æ–¥–∞ —Ä–µ–ª–∞–∫—Å–∞—Ü–∏–∏ –æ–±–æ—Å–Ω–æ–≤–∞—Ç—å –≤—ã–±–æ—Ä –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ —Ä–µ–ª–∞–∫—Å–∞—Ü–∏–∏ w.) –°—Ä–∞–≤–Ω–∏—Ç—å —Å–∫–æ—Ä–æ—Å—Ç—å —Å—Ö–æ–¥–∏–º–æ—Å—Ç–∏ –º–µ—Ç–æ–¥–æ–≤.

–£–ö–ê–ó–ê–ù–ò–ï. 

–ü—Ä–∏ –ø—Ä–æ–≤–µ—Ä–∫–µ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ–≥–æ –∏ –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ–≥–æ —É—Å–ª–æ–≤–∏—è —Å—Ö–æ–¥–∏–º–æ—Å—Ç–∏ –∏—Ç–µ—Ä–∞—Ü–∏–æ–Ω–Ω—ã—Ö –º–µ—Ç–æ–¥–æ–≤ –¥–ª—è –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—è —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –º–∞—Ç—Ä–∏—Ü –≤–æ—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å—Å—è –≤—Å—Ç—Ä–æ–µ–Ω–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–µ–π eigenvals(M) –ø–∞–∫–µ—Ç–∞ MATHCAD (–ª–∏–±–æ –¥—Ä—É–≥–æ–π –≤—Å—Ç—Ä–æ–µ–Ω–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–µ–π –≤ wolfram mathematica –∏–ª–∏ –≤ python), –≤–æ–∑–≤—Ä–∞—â–∞—é—â–µ–π —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –º–∞—Ç—Ä–∏—Ü—ã M.

$$n = 20$$

$$m = 1$$

$$–î–æ–ø. –º–µ—Ç–æ–¥ - –ó–µ–π–¥–µ–ª—è$$

#### –ó–∞–¥–∞–¥–∏–º –º–∞—Ç—Ä–∏—Ü—É

```python
n = 20
m = 1
```

```python
A1 = np.zeros((n, n))
b1 = np.zeros(n)

q = 0.993 + ((-1)**m)* m * 10**(-4)
tetta = sum([q**(2*i) for i in range(1, n+1)])
tm = (m + n)
for i in range(n):
    b1[i] = tm * (q)**i
    for j in range(n):
        if i==j:
            A1[i, j] = tetta + 2*(q**i)*tm + (tm)**2
        else:
            A1[i, j] = tetta + (q**i + q**j)*tm
print(A1)
print()
print(b1)
```

```python
B1 = np.zeros_like(A1)
c1 = np.zeros_like(b1)

for i in range(len(B1)):
    for j in range(len(B1)):
        if i != j:
            B1[i][j] = -A1[i][j]/A1[i][i]
    c1[i] = b1[i]/A1[i][i]
print(B1)
print()
print(c1)
```

```python
x_solve2 = np.linalg.solve(A1, b1)
x_solve2
```

```python
np.linalg.norm(B1, np.inf)
```

–î–ª—è —ç—Ç–æ–π –º–∞—Ç—Ä–∏—Ü—ã –º–µ—Ç–æ–¥ —Ä–∞—Å—Ö–æ–¥–∏—Ç—Å—è

```python
def iters_method(B, c, tol):
    n = len(B) #—Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å
    print(n)
    xk = np.ones(n)
    xk1 = np.copy(xk)
    iters = 0
    while True: 
        xk1 = B@xk +c
        iters += 1
        if iters > 1000:
            print(f"–ü—Ä–µ–≤—ã—à–µ–Ω–æ –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–µ –Ω–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏—Ç–µ—Ä–∞—Ü–∏–π ({iters})")
            return xk,iters
        if (np.linalg.norm(xk - xk1, np.inf) < tol):
            return xk,iters
        xk = xk1
    return xk, iters
```

```python
B_test = B
c_test = c
x_solve = x_solve1
```

```python
tol=1e-6
x_z, iters = iters_method(B_test, c_test, tol=tol)
print('–†–µ—à–µ–Ω–∏–µ —Å –ø–æ–≥—Ä–µ—à–Ω–æ—Å—Ç—å—é:',tol)
for i in range(len(x_z)):
    print("x",i+1,' = ', x_z[i],sep='')
print('–ö–æ–ª–≤–æ –∏—Ç–µ—Ä–∞—Ü–∏–π',iters)
print()
abs_err = np.linalg.norm(x_z - x_solve, np.inf)
print('–ê–±—Å–æ–ª—é—Ç–Ω–∞—è –ø–æ–≥—Ä–µ—à–Ω–æ—Å—Ç—å: ', abs_err)
```

–î–ª—è –ø—Ä–∏–º–µ—Ä–∞ —Ä–∞–±–æ—Ç—ã –±—ã–ª–∞ –≤–∑—è—Ç–∞ –º–∞—Ç—Ä–∏—Ü–∞ –∏–∑ –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ –∑–∞–¥–∞–Ω–∏—è.

```python
tol=1e-6
x_z, iters = zeidel_tol(B_test, c_test, tol=tol,it=200)
print('–†–µ—à–µ–Ω–∏–µ —Å –ø–æ–≥—Ä–µ—à–Ω–æ—Å—Ç—å—é:',tol)
for i in range(len(x_z)):
    print("x",i+1,' = ', x_z[i],sep='')
print('–ö–æ–ª–≤–æ –∏—Ç–µ—Ä–∞—Ü–∏–π',iters)
print()
abs_err = np.linalg.norm(x_z - x_solve, np.inf)
print('–ê–±—Å–æ–ª—é—Ç–Ω–∞—è –ø–æ–≥—Ä–µ—à–Ω–æ—Å—Ç—å: ', abs_err)
```
