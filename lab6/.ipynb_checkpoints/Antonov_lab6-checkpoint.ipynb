```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([0, 0.375, 0.563, 0.75, 1.125, 1.313, 1.5, 1.69, 1.875, 2.063, 2.25,2.438,2.625,2.813,3])
y = np.array([4.568,3.365,2.81,2.624,0.674,0.557,0.384,-0.566,-1.44,-1.696,-1.91,-2.819,-3.625,-3.941,-4.367])
```

```python
def P(a, m, t):
    return sum(a[i]*t**i for i in range(m + 1))
```

```python
def sigma(a, n, m):

    return np.sqrt(sum((P(a, m, x[i]) - y[i])**2 for i in range(n) ) /(n-m))
```

```python
def mnk(x, y, m):

    n = len(x) - 1
    b = np.zeros(m + 1)
    G = np.zeros((m + 1, m + 1))
    a = np.zeros(m + 1)
    for j in range(m + 1):
        b[j] = sum(y[i]*(x[i])**j for i in range(n + 1))
        for k in range(m + 1):
            G[j, k] = sum((x[i])**(j+k) for i in range(n + 1))
    a = np.linalg.solve(G, b)
    return a
```

```python
n = len(x)
deg = np.zeros(n)
for m in range(len(x)):
    a = mnk(x, y, m)
    deg[m] = sigma(a, n, m)
deg
```

```python
index = np.arange(len(deg))
plt.bar(index, deg);
```

# Задание 6.3.1


$a e^{bx^2}$

```python
x = np.array([-2.5,-2,-1.5,-1,-0.5,0,0.5,1,1.5,2,2.5])
y = np.array([0.875,0.29523,0.75958,1.49184,2.23671,2.56,2.23671,1.49184,0.75958,0.29523,0.0876])

# применяем замену переменных: x' = ln(x), y' = y
x_new = np.power(x,2)
y_new = np.log(y)

# вычисляем параметры a и b методом наименьших квадратов
n = len(x_new)
sum_x = np.sum(x_new)
sum_y = np.sum(y_new)
sum_xy = np.sum(x_new*y_new)
sum_x2 = np.sum(x_new**2)

a = (sum_xy - sum_x*sum_y/n) / (sum_x2 - sum_x**2/n)
b = sum_y/n - a*sum_x/n

# выводим результаты
print("Параметры модели:")
print("a =", a)
print("b =", b)

# проверяем результаты, возвращаясь к исходным переменным
b_orig = b
a_orig = np.exp(a)
print("Параметры модели в исходных переменных:")
print("a =", a_orig)
print("b =", b_orig)
```

# Задание 6.7.1

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = -np.pi
b = np.pi
k = 10
```

```python
def y(x):
    return np.abs(np.sin(x))
```

```python
x_points = np.linspace(a, b, k)
x_points
```

```python
y_points = y(x_points)
y_points
```

```python
# функция, вычисляющая значение интерполяционного многочлена 1-ой степени по точкам
def interpolate(x, x0, y0, x1, y1):
    a1 = (y1-y0)/(x1-x0)
    a0 = y0 - a1 * x0
    return a0 + a1*x
```

```python
x = np.linspace(a, b, 3*k)
x
```

```python
y_linear = np.zeros(len(x))
```

```python
for i in range(k-1):
    for j in range(3):
        idx = i*3 + j
        xj = x_points[i] + (j/2)
        y_linear[idx] = interpolate(xj, x_points[i], y_points[i], x_points[i+1], y_points[i+1])
```

```python
from scipy import interpolate
f_interp = interpolate.interp1d(x_points, y_points)
y_global = f_interp(x)
y_global
```

```python
plt.scatter(x_points, y_points, label="Точки интерполяции")
plt.plot(x, y_linear, label='Кусочно-линейная интерполяция')
plt.plot(x, y_global, label='Глобальная интерполяция')
plt.legend()
plt.show()
```

```python
delta_linear = np.abs(y(x) - y_linear)
delta_global = np.abs(y(x) - y_global)
```

```python
delta_linear
```

```python
plt.plot(x,delta_linear, label='Погрешность кусочно-линейной интерполяции')
plt.plot(x,delta_global, label='Погрешность глобальной интерполяции')
plt.legend()
```

# Задание 6.9.1

```python
from scipy.interpolate import CubicSpline
```

```python
#Зададим функцию, которую нужно приблизить:
def f(x):
    return 5*np.sin(x**2)
```

```python
def lagrange(x, y, z):
    n = len(x)
    L = np.zeros_like(z)
    for k in range(n):
        p = np.ones_like(z)
        for j in range(n):
            if j != k:
                p *= (z - x[j]) / (x[k] - x[j])
        L += y[k] * p
    return L
```

```python
#Зададим отрезок, на котором будем строить приближение, и количество узлов интерполяции:
a = 0
b = 4
n = 11 # количество узлов интерполяции

x = np.linspace(a, b, 1000)
plt.plot(x, f(x),label="f(x)")

x_interp = np.linspace(a, b, n)
y_interp = f(x_interp)
y_approx = lagrange(x_interp, y_interp, x)
plt.plot(x, y_approx,label='interpolation');

cs = CubicSpline(x_interp, y_interp, bc_type='clamped')
y_spline = cs(x)
plt.plot(x, y_spline, label='spline')

plt.legend()
plt.show()
```
